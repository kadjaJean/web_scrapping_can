```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
# Liste des liens
#liste_liens = ["lien1", "lien2", "lien3"]  # Remplacez cela par votre propre liste de liens

# Initialisation du navigateur
browser = webdriver.Edge()

# Dictionnaire pour stocker les données
donnees = {"URL": [], "Title": [], "Subscribers": [], "Views": [], "Likes": [], "Publication_date": [],"Duration":[],"Comments":[],"Description":[] }

try:
    for i in liste_liens:
        browser = webdriver.Edge()
        time.sleep(10)
        browser.get(i)
        time.sleep(10)

        # Ajoutez un délai d'attente si nécessaire ici
        # WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "imso_mh__mh-bma")))

        l = browser.find_element(By.XPATH, '//*[@id="liveresults-sports-immersive__match-fullpage"]/div/div[2]/div[4]/div[1]/div[3]/div/div/div/div[2]')
        outer_html = l.get_attribute('outerHTML')

        # Utilisez BeautifulSoup sur outer_html
        inf = BeautifulSoup(outer_html, 'html.parser')
        equi = inf.find_all('a')
        h = equi[0]
        a1 = h["href"]
        if "youtube" not in a1:
            h = equi[1]
            a1 = h["href"]
        
        # browser_1=webdriver.Edge()
        
        donnees["URL"].append(a1)
        time.sleep(10)
        browser.get(a1)
        time.sleep(10)
        #l1_a = browser.find_element(By.CLASS_NAME, "style-scope ytd-watch-flexy")
        #outer_html = l1_a.get_attribute('outerHTML')
        l1_a = browser.find_element(By.XPATH, '//*[@id="primary-inner"]')
        outer_html = l1_a.get_attribute('outerHTML')
        inf_b = BeautifulSoup(outer_html, 'html.parser')
        
        # descri
        
        k=inf_b.find_all('yt-attributed-string')
        desc=k[2].text
        desc1=desc.replace("\xa0\xa0\n","")
        desc2=desc1.replace("\xa0","")
        desc3=desc2.replace("\n","")
        donnees["Description"].append(desc3)
        
        # c0mments
        l = browser.find_elements(By.XPATH, ' //*[@id="title"]')
        # j must be a l'intereieur
        j=0
        listes=[]
        for i in l:
            outer_html = i.get_attribute('outerHTML')
            in8 = BeautifulSoup(outer_html, 'html.parser')
            if in8.find_all('yt-formatted-string'):
                if len(in8.find_all('yt-formatted-string'))>=1:
                    if in8.find_all('yt-formatted-string')[0].text=='Commentaires':
                        listes.append(j)
        
            j=j+1
        outer_html = l[listes[0]].get_attribute('outerHTML')
        in8 = BeautifulSoup(outer_html, 'html.parser')
        coment=in8.find_all('yt-formatted-string')
        coment1=coment[1].text
        donnees["Comments"].append(coment1)
        
        
        #duree
        
        dure=inf_b.find_all('span', {'class': 'ytp-time-duration'})
        duree=dure[0].text
        donnees["Duration"].append(duree)
        j1 = inf_b.find_all('h1')
        k1 = j1[0]
        titre = k1.get_text(strip=True)
        donnees["Title"].append(titre)
        
        # nbre ab0nnes
        k8 = inf_b.find_all('yt-formatted-string', id='owner-sub-count')
        k9 = k8[0].text
        texte_sans_espace_insecable = k9.replace('\xa0', ' ')
        nb_abonne = texte_sans_espace_insecable.strip()
        donnees["Subscribers"].append(nb_abonne)
        
        # nbre vues
        test = inf_b.select('yt-formatted-string>span')
        nb_vues = test[4].text
        nb_vues1 = nb_vues.replace("\xa0", " ")
        donnees["Views"].append(nb_vues1)
        
        # nbre likes
        v1 = inf_b.find_all('yt-formatted-string')
        texte_complet = v1[5].text
        texte_sans_espace_insecable = texte_complet.replace('\xa0', ' ')
        nb_likes = texte_sans_espace_insecable
        donnees["Likes"].append(nb_likes)
        
        # date
        day_mth = v1[7].text
        year = v1[8].text
        date = day_mth + year
        donnees["Publication_date"].append(date)

finally:
    browser.quit()
    time.sleep(10)
```

```python
# boucle
donnees = {"Club": [], "MJ": [], "G": [], "N": [], "P": [], "BP": [],"BC":[],"DB":[],"Pts":[] }
club=["Sénégal","Cameroun","Guinée","Gambia"]
for i in club:
    donnees['Club'].append(i)
    donnees['MJ'].append(df[(df["Team_1"]==i) | (df["Team_2"]==i)].shape[0])
    donnees['G'].append((df[(df["Team_1"]==i) & (df["Goal_Team_1"]>df["Goal_Team_2"])].shape[0]) + (df[(df["Team_2"]==i) & (df["Goal_Team_1"]<df["Goal_Team_2"])].shape[0]))
    g=(df[(df["Team_1"]==i) & (df["Goal_Team_1"]>df["Goal_Team_2"])].shape[0]) + (df[(df["Team_2"]==i) & (df["Goal_Team_1"]<df["Goal_Team_2"])].shape[0])
    donnees['N'].append((df[(df["Team_1"]==i) & (df["Goal_Team_1"]==df["Goal_Team_2"])].shape[0]) + (df[(df["Team_2"]==i) & (df["Goal_Team_1"]==df["Goal_Team_2"])].shape[0]))
    n=(df[(df["Team_1"]==i) & (df["Goal_Team_1"]==df["Goal_Team_2"])].shape[0]) + (df[(df["Team_2"]==i) & (df["Goal_Team_1"]==df["Goal_Team_2"])].shape[0])
    donnees['P'].append((df[(df["Team_1"]==i) & (df["Goal_Team_1"]<df["Goal_Team_2"])].shape[0]) + (df[(df["Team_2"]==i) & (df["Goal_Team_1"]>df["Goal_Team_2"])].shape[0]))
    donnees['BP'].append((df[(df["Team_1"]==i)]["Goal_Team_1"].sum()) + (df[df["Team_2"]==i]["Goal_Team_2"].sum()))
    bp=(df[(df["Team_1"]==i)]["Goal_Team_1"].sum()) + (df[df["Team_2"]==i]["Goal_Team_2"].sum())
    donnees['BC'].append((df[(df["Team_1"]==i)]["Goal_Team_2"].sum()) + (df[df["Team_2"]==i]["Goal_Team_1"].sum()))
    bc=(df[(df["Team_1"]==i)]["Goal_Team_2"].sum()) + (df[df["Team_2"]==i]["Goal_Team_1"].sum())
    donnees['DB'].append(bp-bc)
    donnees['Pts'].append(3*g+n)
    
```
